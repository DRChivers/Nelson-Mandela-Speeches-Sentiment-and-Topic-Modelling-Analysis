```python
import os
import requests

base_url = "https://atom.nelsonmandela.org/index.php/za-com-mr-s-{};ead?sf_format=xml"

# check if the output folder exist if not create it
if not os.path.exists("output"):
    os.makedirs("output")

for i in range(1, 1598):
    url = base_url.format(i)
    response = requests.get(url)
    if response.ok:
        xml_data = response.text
        # Write the XML data to a file in the output folder
        with open(f"output/xml_{i}.xml", "wb") as xml_file:
            xml_file.write(xml_data.encode("utf-8"))
        print(f"File number {i} has been downloaded")
    else:
        print(f"{url} is not accessible, skipping....")
```
